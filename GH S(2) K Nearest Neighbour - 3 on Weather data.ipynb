```python
import os
import pandas as pd 
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
```

```python
os.chdir('C:\\Users\\ABC\\6. Datasets')
```

```python
df = pd.read_csv('weather.csv')
```

```python
df.dropna(inplace= True)
df.head()
```

```python
df['humidity_val'] = (df['relative_humidity_3pm']>25)*1
df
```

```python
df.drop('relative_humidity_3pm', axis = 1, inplace = True)
df
```

```python
y = df['humidity_val']
x = df.drop('humidity_val', axis = 1)
```

```python
for k in range(707,710):
    train_x, test_x, train_y, test_y = train_test_split(x,y,test_size=0.20, random_state=k)
    knn = KNeighborsClassifier(n_neighbors=7)
    model = knn.fit(train_x, train_y)
    y_pred = knn.predict(test_x)
    acc = accuracy_score(y_pred, test_y)
    if acc> 0.858:
        print(k, acc)
```
